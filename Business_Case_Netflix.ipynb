```python
import pandas as pd

# Assuming 'data' is your DataFrame with 'type' and 'duration' columns

# Function to create separate columns out of 'duration' column based on 'type'
def split_duration(row):
    if row["type"] == "Movie" and pd.notna(row['duration']):
        return int(row['duration'].split(' ')[0])
    elif row['type'] == "TV Show" and pd.notna(row['duration']):
        return int(row['duration'].split(' ')[0])
    else:
        return None  # or any default value you prefer

# Apply the function to create a new column 'duration_in_minutes'
data['duration_in_minutes'] = data.apply(split_duration, axis=1)

# Extract number of seasons for TV shows
data['num_seasons'] = data.apply(lambda row: int(row['duration'].split(' ')[0])
if pd.notna(row['duration']) and row['type'] == 'TV Show' else None, axis=1)

# Set duration_in_minutes to 0 for TV shows
data.loc[data['type'] == "TV Show", "duration_in_minutes"] = 0

# Display the DataFrame
data[["show_id", "type", "title", "duration", "duration_in_minutes", "num_seasons"]].head(10)

```

## **Univariate Analysis separately for shows and movies**

```python
data['date_added'] = pd.to_datetime(data['date_added'])

# Extract month and year
data['month_added'] = data['date_added'].dt.month
data['year_added'] = data['date_added'].dt.year

# Filter TV shows and movies
tv_shows = data[data['type'] == 'TV Show']
movies = data[data['type'] == 'Movie']
```

```python
# Mapping numeric months to month names
month_mapping = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

data['month_added'] = data['month_added'].map(month_mapping)
```

```python
# Filter TV shows and movies
tv_shows = data[data['type'] == 'TV Show']
movies = data[data['type'] == 'Movie']

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Plot for TV shows
plt.subplot(1, 2, 1)
sns.countplot(y='month_added', data=tv_shows, hue='month_added', palette='viridis', order=list(month_mapping.values()), legend=False)
plt.title('Distribution of TV Shows Added per Month')

# Plot for movies
plt.subplot(1, 2, 2)
sns.countplot(y='month_added', data=movies, hue='month_added', palette='mako', order=list(month_mapping.values()), legend=False)
plt.title('Distribution of Movies Added per Month')

plt.tight_layout()
plt.show()
```

**`TV Shows:`**

The highest number of TV shows is added in December, followed by July and September.
The lowest number of TV shows is added in February.


**`Movies:`**

The highest number of movies is added in July , followed by April and October.
The lowest number of movies is added in February .


## *Observations:*

For TV shows, December has the highest count, suggesting it might be a popular month for adding TV show content to the platform, possibly due to holidays and increased viewership during that time.
For movies, July has the highest count, indicating a preference for adding movies during the summer months, possibly aligning with the vacation season and increased leisure time.
In summary, December is preferable for TV shows, while July is preferable for movies based on the given data.

## **best week for release analysis**

```python
# Convert 'date_added' column to datetime format
data['date_added'] = pd.to_datetime(data['date_added'])

# Extract month and week from 'date_added' column
data['week'] = data['date_added'].dt.strftime('%U-%B')
data['week'] = data['date_added'].dt.isocalendar().week
```

```python
# Filter data for movies only
movies_data = data[data['type'] == 'Movie']

# Group by month and week
grouped_data = movies_data.groupby(['week'])

# Calculate the count of movies added for each month and week
week_counts = grouped_data.size().fillna(0)

# Identify the top 10 weeks
top_10_weeks = week_counts.nlargest(10)

# Display the insights
print("Top 10 Weeks for Movie Releases:")
print(top_10_weeks)
```

```python
# Filter data for TV shows only
tv_shows_data = data[data['type'] == 'TV Show']

# Group by month and week
grouped_data_tv = tv_shows_data.groupby(['week'])

# Calculate the count of TV shows added for each month and week
week_counts_tv = grouped_data_tv.size().fillna(0)

# Identify the top 10 weeks for TV shows
top_10_weeks_tv = week_counts_tv.nlargest(10)

# Display the insights for TV shows
print("Top 10 Weeks for TV Show Releases:")
print(top_10_weeks_tv)
```

## **TOP 10 GENRES CORRESPONDS TO TOP 10 COUNTRIES**

```python
data.head()
```

```python
# Count the number of titles per country
country_counts = data['country'].value_counts().head(10)

# Filter data for the top 10 countries
top_10_countries_data = data[data['country'].isin(country_counts.index)]

# Exploding the 'listed_in' column to create multiple rows for each genre
data_exploded = top_10_countries_data.explode('listed_in')

# Group by country and genre, then count occurrences
country_genre_counts = data_exploded.groupby(['country', 'listed_in']).size().reset_index(name='count')
```

```python
colors = plt.cm.viridis(np.linspace(0, 1, 10))   # Create a color palette for all genres
fig, axes = plt.subplots(5, 2, figsize=(18, 24))  # Create subplots

# Iterate over each country
for i, country in enumerate(country_counts.index):
    if i >= 10:
        break

    # Filter data for the current country
    country_data = country_genre_counts[country_genre_counts['country'] == country]
    # Sort by count in descending order
    country_data = country_data.sort_values(by='count', ascending=False)
    # Take top 10 genres
    top_10_genres = country_data.head(10)

    # Plotting
    ax = axes[i // 2, i % 2]  # Get the current subplot
    bars = ax.barh(top_10_genres['listed_in'], top_10_genres['count'], color=colors)
    ax.set_xlabel('Number of Titles')
    ax.set_ylabel('Genre')
    ax.set_title(f'Top 10 Genres in {country}')
    ax.invert_yaxis()  # Invert y-axis to have the highest count at the top
    ax.grid(False)  # Turn off the grid


# Adjust layout to prevent overlap of labels
plt.tight_layout()
plt.show()
```

```python
# Create subplots for 5 rows and 2 columns
fig, axes = plt.subplots(5, 2, figsize=(24, 30))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Iterate over each country
for i, country in enumerate(country_counts.index):
    if i >= 10:
        break

    # Filter data for the current country
    country_data = country_genre_counts[country_genre_counts['country'] == country]
    # Sort by count in descending order
    country_data = country_data.sort_values(by='count', ascending=False)
    # Take top 10 genres
    top_10_genres = country_data.head(10)

    # Plotting
    ax = axes[i]
    ax.pie(top_10_genres['count'], labels=top_10_genres['listed_in'], autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Top 10 Genres in {country}')

# Adjust layout
plt.tight_layout()
plt.show()
```
